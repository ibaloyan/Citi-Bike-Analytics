```python
# NYC Citibike Program 2018 - get 2018June-August tripdata
import pandas as pd
import datetime
import numpy as np
import os
```

```python
# Get NYC Citibike tripdata csv files from 201806 to 201808
filepath_jun = os.path.join("Resources" ,"JC-201806-citibike-tripdata.csv")
filepath_jul = os.path.join("Resources" ,"JC-201807-citibike-tripdata.csv")
filepath_aug = os.path.join("Resources" ,"JC-201808-citibike-tripdata.csv")

nycbike_2018jun = pd.read_csv(filepath_jun)
nycbike_2018jul = pd.read_csv(filepath_jul)
nycbike_2018aug = pd.read_csv(filepath_aug)
```

```python
nycbike_2018jun.head(10)
```

```python
nycbike_jun_jul_aug = pd.concat([nycbike_2018jun,nycbike_2018jul,nycbike_2018aug])
```

```python
# Output concatenated 2018 July-Aug citibike tripdata to CSV file 
filepath_2018_jun_aug = os.path.join("Resources" ,"JC-2018-jun-aug-citibike-tripdata1.csv")
nycbike_jun_jul_aug.to_csv(filepath_2018_jun_aug, index=False)
nycbike_jun_jul_aug.tail(10)
```
